```python
import requests 
import pandas as pd
```

```python
def test_artist_sales(artist_address, params={}):
    result = requests.get(f'http://localhost:8000/artist/{artist_address}/sales')
    result = result.json()
    columns, data = result['columns'], result['data']
    df = pd.DataFrame(data, columns=columns)
    df['timestamp'] = df['timestamp'].apply(pd.to_datetime)
    return df 
```

```python
xcopy_address = '0x39Cc9C86E67BAf2129b80Fe3414c397492eA8026'
df = test_artist_sales(xcopy_address)
df.head()
```
